NumPy Exercises for Data Analysis

```python
# 1. Import numpy as np and see the version
import numpy as np
print("Versión de NumPy:", np.__version__)
```

```python
# 2. Create a 1D array of numbers from 0 to 9
arr = np.arange(10)
print("\nEjercicio 2 - Array del 0 al 9:")
print(arr)
```

```python
# 3. Create a 3×3 numpy array of all True’s
arr_true = np.full((3, 3), True, dtype=bool)
print("\nEjercicio 3 - Array 3x3 de True's:")
print(arr_true)
```

```python
# 4. Extract all odd numbers from arr
arr = np.arange(10)
arr_impares = arr[arr % 2 == 1]
print("\nEjercicio 4 - Números impares:")
print(arr_impares)
```

```python
# 5. Replace all odd numbers in arr with -1
arr = np.arange(10)
arr[arr % 2 == 1] = -1
print("\nEjercicio 5 - Array con impares reemplazados por -1:")
print(arr)
```

```python
#  6. Convert a 1D array to a 2D array with 2 rows
arr = np.arange(10)
arr_2d = arr.reshape(2, -1)  # -1 calcula automáticamente las columnas
print("\nEjercicio 6 - Array 1D convertido a 2D (2 filas):")
print(arr_2d)
```

```python
#  7. Stack arrays a and b vertically
a = np.arange(10).reshape(2, -1)
b = np.repeat(1, 10).reshape(2, -1)
resultado_vertical = np.vstack([a, b])
# Alternativa: resultado_vertical = np.concatenate([a, b], axis=0)
print("\nEjercicio 7 - Arrays apilados verticalmente:")
print(resultado_vertical)
```

```python
#  8. Stack the arrays a and b horizontally.
a = np.arange(10).reshape(2, -1)
b = np.repeat(1, 10).reshape(2, -1)
resultado_horizontal = np.hstack([a, b])
# Alternativa: resultado_horizontal = np.concatenate([a, b], axis=1)
print("\nEjercicio 8 - Arrays apilados horizontalmente:")
print(resultado_horizontal)
```

```python
#  9. Get all items between 5 and 10 from a.
# your code here
import numpy as np
a = np.arange(15)
resultado = a[(a >= 5) & (a <= 10)]
print("\nEjercicio 9 - Elementos entre 5 y 10:")
print(resultado)
```

```python
#  11. Swap columns 1 and 2 in the array arr.
# Hint
# Desired Output:
# [[0 2 1]
#  [3 5 4]
#  [6 8 7]]


arr = np.arange(9).reshape(3,3)
print(arr)
# your code here
arr = np.arange(9).reshape(3, 3)
print("\nEjercicio 11 - Array original:")
print(arr)
arr[:, [1, 2]] = arr[:, [2, 1]]
print("Array con columnas 1 y 2 intercambiadas:")
print(arr)
```
